#### Connect with OpenAI Model

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")

res = llm.invoke("Who is the president of the India?")
print(res.content)
```

#### Connect with Google Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")

res = llm.invoke("Who is the president of the India?")
print(res.content)
```

#### Static Prompts

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")

prompts = [
    ("system", "You are a Java Developer."),
    ("user", "How to sort an array?"),
]

res = llm.invoke(prompts)
print(res.content)
```

#### Prompt Templates - Dynamic prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatOpenAI(model="gpt-4o")

prompt = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator and translate the text to {language}."},
    {"role": "user", "content": "{query}"},
])

# Option 1
# final_prompt = prompt.format(language="Hindi", query="What is your name?")
# res = llm.invoke(final_prompt)

# Option 2
# res = llm.invoke(prompt.invoke({"language": "Hindi", "query": "What is your name?"}))

# Option 3
chains = prompt | llm
res = chains.invoke({"language": "Hindi", "query": "What is your name?"})

print(res.content)
```
