#### Static Prompts

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")

prompts = [
    ("system", "You are a Java Developer."),
    ("user", "How to sort an array?"),
]

res = llm.invoke(prompts)
print(res.content)
```

#### Prompt Templates - Dynamic prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOpenAI(model="gpt-4o")
out = StrOutputParser()

# My Self Chain Node
def transform_case(result: str):
    return result.upper()

prompt = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator and translate the input text to {language}."},
    {"role": "user", "content": "{query}"},
])

# Option 1
# final_prompt = prompt.format(language="Hindi", query="What is your name?")
# res = llm.invoke(final_prompt)

# Option 2
# res = llm.invoke(prompt.invoke({"language": "Hindi", "query": "What is your name?"}))

# Option 3
chains = prompt | llm | out | transform_case
res = chains.invoke({"language": "Hinglish", "query": "What is your name?"})

print(res)
```
