```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a: int, b: int) -> int:
    """
    Add two numbers together.
    
    Args:
        a : Number One
        b : Number Two
    """
    return a + b


@tool
def multiply_numbers(a: int, b: int) -> int:
    """
    Multiply two numbers.
    
    Args:
        a : Number One
        b : Number Two
    """
    return a * b
```

```python
add_numbers.invoke({ "a": 30, "b": 5})
```

```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python
llm = ChatGroq(model="openai/gpt-oss-120b")

agent = create_agent(
    model=llm,
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a math teacher, and also use tools for calculation."
)
```

```python
response=agent.invoke({"messages": [{"role": "user", "content": "What is 30 plus 5, and then multiply the result by 2?"}]})
print(response["messages"][-1].content)

# for res in response["messages"]:
#     print(res)
#     print("\n\n")
```
